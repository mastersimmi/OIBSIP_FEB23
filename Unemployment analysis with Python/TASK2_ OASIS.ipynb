# 4. Data Visualization    / Exploratory Data Analysis (EDA): 

## HEAT MAP 

```python
fig, ax = plt.subplots(figsize = (10, 7))
sns.heatmap(df.corr(), annot=True, cmap="RdPu").set(title='Heatmap of Correlated values for "Unemployment_Rate_upto_11_2020" ')
plt.show()
```

## PIE PLOTS

```python
#PARTITION ACCORDING TO AREA on dataset
fig, ax = plt.subplots(figsize =(5, 4))
#labels
m=['Rural Areas','Urban Areas']
#title for chart
ax.set_title("\nDistribution of Unemployment Across Rural and Urban Areas in India",color="navy")
# Creating color parameters
colors = ( "turquoise", "lavender")
# Creating explode data
explode = (0.2, 0.0)
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "grey" }
# Creating plot
wedges, texts, autotexts = ax.pie(data_set['Area'].value_counts(),
                                  autopct = "%0.2f",
                                  explode = explode,
                                  labels = m,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 145,
                                  wedgeprops = wp,
                                  textprops = dict(color ="navy"))
# Adding legend
ax.legend(wedges,m,
          title ="Area Types",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")

plt.show()
```

### PARTITION ACCORDING TO REGION DIRECTIONS IN dataset

```python
df['Region.1'].unique()
```

```python
fig, ax = plt.subplots(figsize =(5, 4))

# Labels
regions = ['South', 'Northeast', 'East', 'West', 'North']

# Title for chart
ax.set_title("Unemployment by Region in India", color="purple")

# Creating color parameters
colors = ("peachpuff", "turquoise", "burlywood", "gainsboro", "lavender")

# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0)

# Wedge properties
wp = { 'linewidth': 1, 'edgecolor': "darkgreen" }

# Creating plot
wedges, texts, autotexts = ax.pie(df['Region.1'].value_counts(),
                                  autopct = "%0.2f",
                                  explode = explode,
                                  labels = regions,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="purple"))

# Adding legend
ax.legend(wedges, regions,
          title = "Region",
          loc = "center left",
          bbox_to_anchor = (1.2, 0, 0.5, 1))

plt.setp(autotexts, size = 8, weight = "bold")

plt.show()
```

Unemployment according to Dates in different Regions Directions -

```python
#SCATTER PLOTING
fig = plt.figure(figsize = (9, 6))
sns.scatterplot(x=' Date', y=' Estimated Unemployment Rate (%)',data=df, hue="Region.1")
plt.title('Unemployment Rate by Dates in different Regions')
plt.xticks(rotation=90)
plt.show()
```

```python
#BAR PLOTING
fig = plt.figure(figsize = (9, 6))
sns.barplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=df, hue="Region.1")
plt.title('Unemployment according to Dates in different Regions')
plt.xticks(rotation=90)
plt.show()
```

## Histogram plots

```python
fig = plt.figure(figsize = (9, 6))
sns.histplot(df[' Estimated Unemployment Rate (%)'], color='blue', edgecolor='black')
plt.title('Histogram of Estimated Unemployment Rate (%)')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()
```
